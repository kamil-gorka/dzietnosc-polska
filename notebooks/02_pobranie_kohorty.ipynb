```python
import requests

BASE = "https://bdl.stat.gov.pl/api/v1"

def szukaj_tematow(fraza):
    r = requests.get(
        f"{BASE}/subjects/search",
        params={"name": fraza, "format": "json", "page-size": 50},
    )
    r.raise_for_status()
    for s in r.json()["results"]:
        print(s["id"], "|", s["name"])

szukaj_tematow("ludność")
```

```python
def zmienne_tematu(subject_id):
    r = requests.get(
        f"{BASE}/variables",
        params={"subject-id": subject_id, "format": "json", "page-size": 100},
    )
    r.raise_for_status()
    for v in r.json()["results"]:
        opis = " | ".join(
            str(v[k]) for k in ("n1", "n2", "n3") if v.get(k)
        )
        print(v["id"], "->", opis)

zmienne_tematu("P3968")
```

```python
szukaj_tematow("wieku rozrodczym")
print("---")
szukaj_tematow("płodności")
print("---")
szukaj_tematow("wg wieku matki")
```

```python
def podtematy(parent_id=None):
    params = {"format": "json", "page-size": 100}
    if parent_id:
        params["parent-id"] = parent_id
    r = requests.get(f"{BASE}/subjects", params=params)
    r.raise_for_status()
    for s in r.json()["results"]:
        print(s["id"], "|", s["name"], "| dzieci:", s.get("hasDynamicData"))

podtematy()
```

```python
import pandas as pd

def pobierz_zmienna(var_id, rok_od=1995, rok_do=2025):
    lata = [("year", r) for r in range(rok_od, rok_do + 1)]
    r = requests.get(
        f"{BASE}/data/by-variable/{var_id}",
        params=[("format", "json"), ("unit-level", "0"), ("page-size", "100")] + lata,
    )
    r.raise_for_status()
    wyniki = r.json()["results"][0]["values"]
    return pd.Series(
        {w["year"]: w["val"] for w in wyniki},
        name=var_id,
    )

test = pobierz_zmienna(72299)
print(test.head())
print("lata:", test.index.min(), "-", test.index.max(), "| liczba:", len(test))
```
